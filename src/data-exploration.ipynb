```python
import os
import numpy as np
from tqdm import tqdm
```

```python
BASE_DIR = "../data/comma.ai/comma_research_dataset/comma-dataset/comma-dataset"
mode = "train"
token_dir = os.path.join(BASE_DIR, "tokens", mode)

token_batches = [os.path.join(token_dir, dirr) for dirr in sorted(os.listdir(token_dir)) if dirr.endswith(".npz")]
token_batches
```

```python
all_tokens = []
indices = []
for batch in tqdm(token_batches):
  data = np.load(batch, allow_pickle=True)
  tokens = data["tokens"]          # shape: (B, Ht, Wt), dtype: uint16
  index = data["index"]
  dataset_idx = data["dataset_idx"]
  inner_idx = data["inner_idx"]
  Ht, Wt = data["token_hw"]      # [Ht, Wt]
  B = tokens.shape[0]
  all_tokens.append(tokens)
  indices.append(np.stack([dataset_idx, inner_idx], axis=1))   # shape: (B, 2)
all_tokens = np.concatenate(all_tokens, axis=0)   # shape: (N, Ht, Wt)
indices = np.concatenate(indices, axis=0)   # shape: (N, 2)
```
